```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
cwd = os.getcwd()
print(cwd)
os.chdir("C:/Users/sanat/Pictures/Black Dresses/")
path = os.getcwd()
image_detector_path = "C:\\Users\\sanat\\Documents\\Github\\Image\\frontal_face_default.xml"
face_cascade = cv2.CascadeClassifier(image_detector_path)
flist = os.listdir(path)
valid_exts = [".jpg",".gif",".png",".tga", ".jpeg"]

for f in flist:
    # add image to path
    print(f)
    full_path = os.path.join(path, f)
    # read in the image
    img_bgr = cv2.imread(full_path)
    # convert image to rgb
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    # get the rgb version of the image
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
    faces = face_cascade.detectMultiScale(img_gray)

    plt.imshow(img_gray, cmap=plt.get_cmap("gray"))
    # get the current axis (retain values)
    im = plt.gca()
    for p in faces:
        im.add_patch(Rectangle((p[0], p[1]), p[2], p[3]
                              , fill=None, alpha=1, edgecolor='green'))
   
        
    plt.show()
```
